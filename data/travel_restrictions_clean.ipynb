```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Loading raw data
path = "Raw/travel_restrictions.csv"
travel_restrictions_raw= pd.read_csv(path) 

travel_restrictions_raw
```

```python
#Rename columns 
travel_restrictions_raw.rename(columns = {'adm0_name':'Country', 'iso3':'Country_id', 
                              'optional2':'Quarantine_type', 'published': 'Quarantine_start_date'}, inplace = True)

#Only keeping relevant columns 
columns_to_keep = ["Country", "Country_id", "Quarantine_type", "Quarantine_start_date"]

Renamed_df = travel_restrictions_raw[columns_to_keep]
Renamed_df 
```

```python
date_list = [str(x).replace('.', '-') for x in Renamed_df['Quarantine_start_date']]
date_list_edited = ['01-01-01' if x == 'nan' else x for x in date_list]
date_list_reedited = ['04-02-20' if x == '04-02-20201' else x for x in date_list_edited]
Renamed_df['Quarantine_start_date'] = date_list_reedited
```

```python
# changing start date to datetime object
Renamed_df['Quarantine_start_date'] = pd.to_datetime(Renamed_df['Quarantine_start_date'])
print(Renamed_df['Quarantine_start_date'].dtype)
Renamed_df.head()
```

```python
#Rename NaNs in Quarantine type to "unknown"
non_nans_df= Renamed_df.fillna("Unknown")
non_nans_df
```

```python
#Drop any Country ID's with more than one country ID or starts with x (They are territories)
df = non_nans_df[~non_nans_df["Country_id"].str.contains(",")&~non_nans_df["Country_id"].str.contains("x")]

df
```

```python
df.reset_index(drop= True, inplace =True)
df
```

```python
#Save to output file 
df.to_csv('Clean/travel_restrictions_clean.csv', index=False)
```
